```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#[Action,Adventure,Fantasy,Science Fiction]
import ast
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l

    
```

```python
# cast
import ast
def convert(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i['name'])
            counter=counter+1
        else:
            break
    return l
```

```python
movie2['cast']=movie2['cast'].apply(convert)
```

```python
movie2.head(1)
```

```python
movie2['crew'][0]
```

```python
# directors

import ast

def director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            l.append(i['name'])
            break
    return l
```

```python
movie2['crew']=movie2['crew'].apply(director)
```

```python
movie2['overview']=movie2['overview'].apply(lambda x:x.split())
```

```python
movie2['overview']
```

```python
movie2.head()
```

```python
# removing space

movie2['genres']=movie2['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movie2['keywords']=movie2['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movie2['cast']=movie2['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movie2['crew']=movie2['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movie2.head()
```

```python
# tags

movie2['tags']=movie2['overview']+movie2['genres']+movie2['keywords']+movie2['cast']+movie2['crew']
```

```python
new_df=movie2[['id','title','tags']]
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'][1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
feature_names=cv.get_feature_names_out()
```

```python
feature_names
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
ps.stem('loving')
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y) 



    
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[2]
```

```python
sorted(similarity[0], reverse=True)
```

```python
list(enumerate(similarity[0]))
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('John Carter')
```
